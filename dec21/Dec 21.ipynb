```python
from functools import wraps

import numpy as np
```

```python
all_ops = []

def make_op(f):
    @wraps(f)
    def new_f(regs, *args):
        #regs = np.copy(regs)
        f(regs, *args)
        return regs
    all_ops.append(new_f)
    return new_f

def reg_op(f):
    @wraps(f)
    def new_f(regs, a, b, c):
        return f(regs, a, regs[b], c)
    return new_f

@reg_op
@make_op
def addr(regs, a, b, c):
    regs[c] = regs[a] + b
    
@reg_op
@make_op
def mulr(regs, a, b, c):
    regs[c] = regs[a] * b
    
@reg_op
@make_op
def banr(regs, a, b, c):
    regs[c] = regs[a] & b
    
@reg_op
@make_op
def borr(regs, a, b, c):
    regs[c] = regs[a] | b

@make_op
def addi(regs, a, b, c):
    regs[c] = regs[a] + b
    
@make_op
def muli(regs, a, b, c):
    regs[c] = regs[a] * b
    
@make_op
def bani(regs, a, b, c):
    regs[c] = regs[a] & b
    
@make_op
def bori(regs, a, b, c):
    regs[c] = regs[a] | b

@make_op
def setr(regs, a, b, c):
    regs[c] = regs[a]

@make_op
def seti(regs, a, b, c):
    regs[c] = a
    
@make_op
def gtri(regs, a, b, c):
    regs[c] = 1 if regs[a] > b else 0
    
@make_op
def gtir(regs, a, b, c):
    regs[c] = 1 if a > regs[b] else 0
    
@make_op
def gtrr(regs, a, b, c):
    regs[c] = 1 if regs[a] > regs[b] else 0
    
@make_op
def eqri(regs, a, b, c):
    regs[c] = 1 if regs[a] == b else 0
    
@make_op
def eqir(regs, a, b, c):
    regs[c] = 1 if a == regs[b] else 0
    
@make_op
def eqrr(regs, a, b, c):
    regs[c] = 1 if regs[a] == regs[b] else 0
```

```python
program = [line.strip().split() for line in open('input.txt') if line.strip()]
pc = int(program[0][-1])
program = program[1:]
program = [[x, int(y), int(z), int(a)] for x, y, z, a in program]

regs = np.zeros(8, dtype=int)

counter = 0
while regs[pc] >= 0 and regs[pc] < len(program):
    counter += 1
    if regs[pc] == 28:
        # At this point the next instructions are eqrr 3 0 4, addr 4 5 5, seti 5 8 5
        # So regs[pc] will go to 29 and add regs[4] to the pc. If it adds 0, we go
        # back to the beginning. If it adds 1, then we are out of scope woot
        print("The answer to part 1 is: ", regs[3])
        regs[0] = regs[3]
    if counter % 1000 == 0:
        print('Got to round {}'.format(counter))
    regs = eval(program[regs[pc]][0])(regs, *program[regs[pc]][1:])
    regs[pc] += 1
```
